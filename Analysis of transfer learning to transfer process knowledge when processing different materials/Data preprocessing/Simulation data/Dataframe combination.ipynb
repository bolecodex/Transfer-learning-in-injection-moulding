```python
import numpy as np
import pandas as pd
import pandas as pd
import os
```

```python
Material_list=['APEC','PLEXIGLAS','SABIC','ULTEM','ULTRAMID','VALOX']
for i in Material_list:
    print (i)
```

```python
for i in Material_list:
    filelist=[]# Initialization
    for root, dirs, files in os.walk(".", topdown=False): #find all the txt file in the folder
        for name in files:
            str=os.path.join(root, name)
            if i in str: #Seach all the files whose name contain i
                filelist.append(str) #put all the name of csv files into a list
    SaveFile_Name = r'df_%s.csv'% i  #The name of dataset after combination
    
    df = pd.read_csv('3x2_Lego_APEC1600_SP.csv') #Load the first csv file as initialization
    df=df.drop(index=range(0, len(df)))#Delete all the data, we only need the index
    df.to_csv('./Sorted_DataFrame/%s' % SaveFile_Name,encoding="utf_8_sig",index=False) #Save the first csv file to the material class dataset
    df

    for i in range(0,len(filelist)): #traversal of all the csv file in the filelist, add them to the material class dataset
        print (filelist[i]) 
        df = pd.read_csv(filelist[i],encoding = 'unicode_escape') 
        df.to_csv('./Sorted_DataFrame/%s' % SaveFile_Name,encoding="utf_8_sig",index=False, header=False, mode='a+')
    
    print (pd.DataFrame(filelist))
```
