```python
Example.rename(columns={'Unnamed: 3':'Attribute' }, inplace=True)# Rename the column
```

```python
df_attribute=pd.DataFrame(Example['Attribute'].drop(0))#Get the "Attribute" column
```

```python
filelist=[]# Initialization
for root, dirs, files in os.walk(".", topdown=False): #find all the excel file in the folder
    for name in files:
        str=os.path.join(root, name)
        if str.split('.')[-1]=='xlsx':
            filelist.append(str)
#filelist
```

```python
file_chdir = os.getcwd()# Initialization
file_excel = []
for root, dirs,files in os.walk(file_chdir): #Also find all the excel file in the folder, Alternative
    for file in files:
        if os.path.splitext(file)[1] =='.xlsx':
            file_excel.append(file)
data=df_attribute# Initialization
for file in file_excel:
    print (file.strip('.xlsx')) #print the name of the materials
    df_example=pd.read_excel(file) #Load the file
    df_example.rename(columns={ '#Copyright Cadflow France & Simcon 2016 -- Version : 0005':'Index', 'Unnamed: 1': file.strip('.xlsx'),'Unnamed: 2 ':'Unit','Unnamed: 3':'Attribute' }, inplace=True) #Rename the table
    df_example=pd.DataFrame(df_example[file.strip('.xlsx')].drop(0))# Remove the "xlsx" in the string, Select the column we want to use
    data=pd.concat([data, df_example], axis=1) # Concate all the data 
data
```

# Removing features with low variance and obviously 

```python
pd.set_option('display.max_rows', None) #Show all rows in pandas
pd.set_option('display.max_columns', None) # Show all columns in pandas
```

```python
data.dropna(axis=0,how='all')  #Remove the NaN row
df_Attribute=data.drop(['Attribute'], axis=1) #Remove the Attribute row
#df_Attribute.dropna(axis=0,how='any')
```

```python
Index_0=df_Attribute.loc[(df_Attribute==0).all(axis=1), :].index.values.tolist() # Find the row index where all values are 0
#df1.ix[~(df_Attribute==0).all(axis=1), :]  # Delet them
Index_0
```

```python
df_null =df_Attribute.isnull().all(axis=1) # Check whether one row filled with NaN or not
Index_NaN=df_null[df_null.values==True].index.values.tolist() # Get the row index where all the values are NaN
Index_NaN
```

```python
Index_NaN.extend(Index_0) #Combine the both row index of 0 and NaN
```

```python
#df_Attribute.drop(Index_NaN).duplicated()#未完成 查看是否有重复
```

```python
df=data.drop(Index_NaN) #Drop these useless row
df.index = range(len(df))# Sort the row index
#df.isnull().sum(axis=1)
```

```python
df=df[~df['Attribute'].isin(['Trade Name','Manufacturer','Sort','Description / Information'])] #Delete the row that are not obviously irrelevant
df=df[~df['Attribute'].isin(['Temperature E Variable //','E Variable //'])]
df=df.drop([3,4,5,6])#Delete 'Sort','Description / Information','Update'
```
